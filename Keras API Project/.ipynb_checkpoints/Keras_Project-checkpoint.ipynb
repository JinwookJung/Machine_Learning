```python
total_acc_avg = df.groupby('total_acc').mean(numeric_only=True)['mort_acc']
```

```python
def fill_mort_acc(total_acc, mort_acc):
    if np.isnan(mort_acc):
        return total_acc_avg[total_acc]
    else:
        return mort_acc
```

```python
df['mort_acc'] = df.apply(lambda x: fill_mort_acc(x['total_acc'],x['mort_acc']),axis=1)
```

```python
# mort_acc is 0
df.isnull().sum()
```

```python
# revol_util and the pub_rec_bankruptcies have missing data points, but they account for less than 0.5% of the total data. 
# remove the rows that are missing those values in those columns
df = df.dropna()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
```

```python
df = df.drop('loan_status',axis=1)
```

```python
X = df.drop('loan_repaid',axis=1).values
```

```python
y = df['loan_repaid'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
# Normalizing the Data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Creating the Model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.constraints import max_norm
```

```python
model = Sequential()
# input layer
model.add(Dense(78,  activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(39, activation='relu'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(19, activation='relu'))
model.add(Dropout(0.2))

# output layer
model.add(Dense(units=1,activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=25,
          batch_size=256,
          validation_data=(X_test, y_test), 
          )
```

```python
# Model Evaluation
from tensorflow.keras.models import load_model
```

```python
# model.save('myfavoritemodel.h5')
```

```python
# Evaluating Model Performance
model.history.history
```

```python
# Evaluating Model Performance
losses = pd.DataFrame(model.history.history)
```

```python
losses[['loss','val_loss']].plot()
```

```python
# predictions from the X_test set and display a classification report and confusion matrix for the X_test set.
from sklearn.metrics import classification_report,confusion_matrix
```

```python
pred = (model.predict(X_test) > 0.5).astype("int32")
```

```python
print(classification_report(y_test,pred))
```

```python
confusion_matrix(y_test,pred)
```
